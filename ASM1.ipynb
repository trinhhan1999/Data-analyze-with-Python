```python
# Câu hỏi 7: Liệt kê số lượng trận đấu của mỗi giải đấu bao gồm cả tên giải đấu, sắp xếp theo thứ tự giảm dần số trận đấu
cau7 = pd.read_sql(
   '''
    select name as league_name, league_id, count(*) as number_match from Match
    left join League on League.id = Match.league_id 
    group by league_id 
    order by count(*) desc
    ''', conn)
cau7
```

```python
# Câu hỏi 8: Liệt kê tổng số bàn thắng của đội nhà và đội khách trong mỗi giải đấu
cau8 = pd.read_sql(
        '''
        select name as league_name, sum(home_team_goal) as Goal_home, sum(away_team_goal) as Goal_away from Match
        left join League on League.id = Match.league_id 
        group by name
        '''
        , conn)
cau8
```

```python
# Câu hỏi 9: Liệt kê thông tin các đội từ bảng “Team”
cau9 = pd.read_sql(
        '''
        select * from Team
        '''
        , conn)
cau9
```

```python
# Câu hỏi 10: Liệt kê 20 đội với số bàn thắng sân nhà cao nhất
cau10 = pd.read_sql(
    '''
    SELECT team_long_name AS Team, SUM(home_team_goal) AS Goal
        FROM Match AS m
        LEFT JOIN Team AS t
        ON m.home_team_api_id = t.team_api_id
        GROUP BY Team
        ORDER BY Goal DESC
        LIMIT 20
    '''
    , conn)
cau10
```

```python
# Câu hỏi 11: Liệt kê 20 đội có số bàn thắng trên sân khách cao nhất
cau11 = pd.read_sql(
    '''
    SELECT team_long_name AS Team, SUM(away_team_goal) AS Goal
        FROM Match AS m
        LEFT JOIN Team AS t
        ON m.away_team_api_id = t.team_api_id
        GROUP BY Team
        ORDER BY Goal DESC
        LIMIT 20
    '''
    , conn)
cau11
```

```python
# Câu hỏi 12: Liệt kê tên các đội bóng và tổng số bàn thắng mỗi đội, sắp xếp theo số lượng giảm dần
cau12 = pd.read_sql_query(
    '''
        SELECT     
                   t.team_long_name AS Team,
                   --m.home_team_goal AS home_goal,
                   --m.away_team_goal AS away_goal,
                   COUNT(CASE WHEN m.home_team_goal > away_team_goal THEN (home_team_goal - away_team_goal) end) AS Won
         FROM Match AS m
         LEFT JOIN Team as t
         ON m.home_team_api_id = t.team_api_id
         GROUP BY Team
         ORDER BY Won desc
    ''', conn)
cau12
```

```python
# Câu hỏi 13: Liệt kê tên các đội bóng (long name) và tổng số trận đấu đội đó tham gia, sắp xếp theo thứ tự giảm dần.
cau13 = pd.read_sql_query(
    '''
        SELECT team_long_name AS Team, count(team_api_id) as number_match
        FROM Team AS t
        LEFT JOIN Match AS m
        ON m.home_team_api_id = t.team_api_id or m.away_team_api_id = t.team_api_id
        GROUP BY Team
        ORDER BY Count(team_api_id) DESC
    ''', conn)
cau13
```

```python
# Câu hỏi 14: Liệt kê số trận thắng, thua và hòa của mỗi đội.
cau14 = pd.read_sql_query(
    '''
        SELECT     
                   t.team_long_name AS Team,
                   --m.home_team_goal AS home_goal,
                   --m.away_team_goal AS away_goal,
                   COUNT(CASE WHEN m.home_team_goal > away_team_goal THEN 'Win' END) AS Won,
                   COUNT(CASE WHEN m.home_team_goal < away_team_goal THEN 'Lost' END) AS Lost,
                   COUNT(CASE WHEN m.home_team_goal = away_team_goal THEN 'Tie' END) AS Tie
         FROM Match AS m
         LEFT JOIN Team as t
         ON m.home_team_api_id = t.team_api_id
         GROUP BY Team
         ORDER BY Team
    ''', conn)
cau14
```

```python
# Câu hỏi 15: Liệt kê 10 đội có số trận thắng nhiều nhất
cau15 = pd.read_sql_query(
    '''
        SELECT     
                   t.team_long_name AS Team,
                   --m.home_team_goal AS home_goal,
                   --m.away_team_goal AS away_goal,
                   COUNT(CASE WHEN m.home_team_goal > away_team_goal THEN 'Win' END) AS Won
         FROM Match AS m
         LEFT JOIN Team as t
         ON m.home_team_api_id = t.team_api_id
         GROUP BY Team
         ORDER BY Won desc
         Limit 10
    ''', conn)
cau15
```

```python
# Câu hỏi 16: Liệt kê tỉ lệ phần trăm về số trận của mỗi giải trên tổng tất cả các trận
cau16 = pd.read_sql(
   '''
    SELECT team_long_name AS Team, (count(team_api_id) * 100.0 / (select count(id) from Match)) as percentage
        FROM Team AS t
        LEFT JOIN Match AS m
        ON m.home_team_api_id = t.team_api_id or m.away_team_api_id = t.team_api_id
        GROUP BY Team
        ORDER BY Count(team_api_id) DESC
    ''', conn)
cau16
```

```python
# Câu hỏi 17: Liệt kê tỉ lệ phần trăm về số bàn ghi được của từng giải so với tổng số bàn của tất cả các trận.
cau17 = pd.read_sql(
    '''
    SELECT  
              name as League_name,
              SUM(home_team_goal) + SUM(away_team_goal) as total_goal
        FROM Match AS m
        LEFT JOIN League AS l
        ON m.league_id = l.id 
        GROUP BY name
        ORDER BY total_goal DESC
    ''', conn)
cau17
```

```python
# Câu hỏi 18: Liệt kê tổng số bàn ghi được của từng giải đấu theo từng mùa giải.
cau18 = pd.read_sql(
    '''
    SELECT 
               name AS league_name,
               season AS Season,
               SUM(home_team_goal) + SUM(away_team_goal) as total_goal 
        FROM Match as m
        LEFT JOIN League AS l
        ON m.league_id = l.id
        GROUP BY Season, league_name
        ORDER BY total_goal desc
    ''', conn)
cau18
```
